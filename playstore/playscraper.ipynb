### Imports

```python
from google_play_scraper import app, reviews_all
from transformers import pipeline 
import pandas as pd
import numpy as np
```

## Data Gathering

```python
# scrape all reviews
htmlpro = reviews_all('com.codeliber.htmlpro',sleep_milliseconds=0, lang='en', country='ng' ) 
```

```python
# save all reviews to a dataframe
df = pd.json_normalize(htmlpro)
```

### Sentiment Analysis

```python
# ensure that 'object' datatype is 'string'
df['content'] = df['content'].astype('str')
```

```python
# use the HuggingFace transformers library to create a sentiment analysis pipeline with a siebert model
sentiment_analysis = pipeline('sentiment-analysis', model = 'siebert/sentiment-roberta-large-english')
```

```python
# perform sentiment analysis on the review contents
df['sentiment_result'] = df['content'].apply(lambda x: sentiment_analysis(x))
```

```python
# assess the dataframe
df.head(10)
```

```python
# extract the sentiment label only to a separate column
df['sentiment_label'] = df['sentiment_result'].apply(lambda x: (x[0]['label']))
# extract the sentiment score only to a separate column
df['sentiment_score'] = df['sentiment_result'].apply(lambda x: (x[0]['score']))
```

```python
# save data to a csv file
df.to_csv('learnhtmlpro-sa.csv', index=False)
```
